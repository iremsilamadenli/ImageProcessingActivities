```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files


imgage = cv2.imread('roses.JPG')

gray = cv2.cvtColor(imgage, cv2.COLOR_BGR2GRAY)

filter1 = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
filter2 = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])

filtered1 = cv2.filter2D(gray, -1, filter1)
filtered2 = cv2.filter2D(gray, -1, filter2)

plt.imshow(filtered1, cmap='jet')
plt.show()

plt.imshow(filtered2, cmap='viridis')
plt.show()

plt.imshow(np.abs(filtered1), cmap='plasma')
plt.show()

plt.imshow(np.abs(filtered2), cmap='magma')
plt.show()
```
