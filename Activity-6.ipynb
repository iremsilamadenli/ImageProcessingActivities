```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('image.jpeg', cv2.IMREAD_GRAYSCALE)

# Computing the 2D FFT of the image
fft = np.fft.fft2(img)

# Shifting zero-frequency component to the center
fft_shift = np.fft.fftshift(fft)

# Visualize the magnitude of the FFT using log transform
magnitude_spectrum = 20 * np.log(np.abs(fft_shift))

# Displaying the original image and magnitude spectrum
plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])


plt.subplot(122), plt.imshow(magnitude_spectrum, cmap='gray')

plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()

# Inversing shift the FFT
fft_ishift = np.fft.ifftshift(fft_shift)

# Converting back to the spatial domain using Inverse FFT
img_back = np.fft.ifft2(fft_ishift)
img_back = np.real(img_back)

# Checking the difference between the original and the reconstructed image
print("Maximum absolute difference between the images: ", np.max(np.abs(img - img_back)))
```
