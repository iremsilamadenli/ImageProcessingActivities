```python
import cv2
from google.colab.patches import cv2_imshow 
# Read the image
img = cv2.imread('image.jpg')

cv2_imshow(img)
# Convert to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
cv2_imshow(gray_img)

# Detect edges using Canny edge detector with threshold1=100 and threshold2=300
edges1 = cv2.Canny(gray_img, 100, 300)

# Display the edge-detected image with threshold1=100 and threshold2=300
cv2_imshow(edges1)

# Detect edges using Canny edge detector with threshold1=10 and threshold2=300
edges2 = cv2.Canny(gray_img, 10, 300)

# Display the edge-detected image with threshold1=10 and threshold2=300
cv2_imshow(edges2)
```
