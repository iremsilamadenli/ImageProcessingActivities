```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt

# Load & Display the original image
img = cv2.imread('snow.png')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.show()

# Converting the image into HSV color space.
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Displaying V channel and its histogram
V = hsv_img[:, :, 2]
plt.imshow(V, cmap='gray')
plt.title(" V Channel")
plt.show()
plt.hist(V.ravel(), 85, [0, 256])
plt.title(" V Channel ")
plt.show()

# Applying histogram equalization to the V channel & Display the photo
equ_V = cv2.equalizeHist(V)
plt.imshow( equ_V , cmap='gray')
plt.title(" V channel after histeq ")
plt.show()

# Display the histogram of the equalized V channel
plt.hist(equ_V.ravel(), 50, [0, 256])
plt.title("V channel after histeq")
plt.show()

# By using H and S channel from the original image and V channel after histogram equalization, go back to RGB color space.
equalized_hsv_img = cv2.merge([hsv_img[:, :, 0], hsv_img[:, :, 1], equ_V])
result = cv2.cvtColor(equalized_hsv_img, cv2.COLOR_HSV2RGB)

# Display the result
plt.imshow(result)
plt.title("Image after histeq")
plt.show()
```
